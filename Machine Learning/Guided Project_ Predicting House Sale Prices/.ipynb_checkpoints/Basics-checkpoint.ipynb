```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import re
import seaborn as sns
```

```python
data = pd.read_csv("AmesHousing.tsv", delimiter='\t')
```

```python
data.head()
```

```python
def transform_features(df, dt):
    m = df.shape[0]
    df_trans = df.copy()
    null_cols = df_trans.isnull().sum()
    
    #drop columns with more than 25% missing values
    df_trans = df_trans[null_cols[null_cols < m * 0.25].index]
    #transform and drop some columns
    df_trans["Years Before Sold"] = df_trans["Yr Sold"] - df_trans["Year Built"]
    df_trans["Years Since Remod"] = df_trans["Yr Sold"] - df_trans["Year Remod/Add"]
    df_trans.drop(["Order", "PID", "Year Built", "Year Remod/Add", "Yr Sold", "Mo Sold", "Yr Sold", "Sale Type", "Sale Condition"], axis=1, inplace=True)
    for col in dt.keys():
        if col in df_trans.select_dtypes(["object"]).columns:
            df_trans[col] = df_trans[col].map(dt[col])
            
    num_cols = df_trans.select_dtypes(include=['int', 'float'])
    num_null = num_cols.isnull().sum()
    df_trans = df_trans.drop(num_null[num_null > m * 0.05].index.tolist(), axis = 1)
    values = {col: num_cols[col].mean() for col in num_cols.columns}
    df_trans = df_trans.fillna(values)
    
    str_cols = df_trans.select_dtypes(include=['object'])
    str_null = str_cols.isnull().sum()
    df_trans = df_trans.drop(str_null[str_null > 0].index.tolist(), axis = 1)
    
    return df_trans
```

```python
def select_features(df):
    for col in dn.keys():
        if col in df.columns:
            if df[col].value_counts().iloc[0] <= len(df) * 0.95 or len(df[col].value_counts()) <= 10:
                df[col] = data_trans[col].astype("category")
                df = pd.concat([df, pd.get_dummies(df[col])], axis=1)
            df = df.drop(col, axis=1)
    return df
```

```python
def train_and_test(df, features, k=0):
    model = LinearRegression()
    if k == 0:
        train = df[:1460]
        test = df[1460:]
        model.fit(train[features], train["SalePrice"])
        test_pred = model.predict(test[features])
        rmse = mean_squared_error(test["SalePrice"], test_pred)**0.5
    else:
        kf = KFold(n_splits=k)
        rmses = []
        for train_idx, test_idx in kf.split(df):
            train = df.iloc[train_idx]
            test = df.iloc[test_idx]            
            model.fit(train[features], train["SalePrice"])
            test_pred = model.predict(test[features])
            rmses.append(mean_squared_error(test["SalePrice"], test_pred)**0.5)
        rmse = np.mean(rmses)
    return rmse
```

Transform ordinal data to integer values.

```python
with open("doc") as d:
    doc = d.read()
lines = doc.split('\n')
dt = {}
dn = {}
for line in lines:
    if re.match(".+\(.+\):.+", line):
        case = 0
        key = line.split('(')[0].strip(' ').strip('\t')
        if line.split('(')[1].split(')')[0] == "Ordinal":
            dt[key] = []
            case = 1
        elif line.split('(')[1].split(')')[0] == "Nominal":
            dn[key] = []
            case = 2
    elif case == 1:
        if re.match(".+\t.+", line):
            dis = line.split('\t')[0].strip(' ')
            dt[key].append(dis)
    elif case == 2:
        if re.match(".+\t.+", line):
            dis = line.split('\t')[0].strip(' ')
            dn[key].append(dis)
for k, v in dt.items():
    dic = {}
    for i in range(len(v)):
        dic[v[i]] = len(v) - i
    dt[k] = dic
```

```python
data_trans = transform_features(data, dt)
```
