```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv("fandango_score_comparison.csv")
movies
```

```python
plt.hist(movies["Metacritic_norm_round"], width=0.3)
plt.xlim(0.0,5.0)
plt.show()
```

```python
plt.hist(movies["Fandango_Stars"], width=0.3)
plt.xlim(0.0,5.0)
plt.show()
```

Fandango Stars are commonly higher than normalized Metacritic scores. And the stars are all above 3.0.

```python
fd_mean = movies["Fandango_Stars"].mean()
fd_med = np.median(movies["Fandango_Stars"])
fd_std = np.std(movies["Fandango_Stars"])
mc_mean = movies["Metacritic_norm_round"].mean()
mc_med = np.median(movies["Metacritic_norm_round"])
mc_std = np.std(movies["Metacritic_norm_round"])
print("fd_mean: "+str(fd_mean))
print("fd_med: "+str(fd_med))
print("fd_std: "+str(fd_std))
print("mc_mean: "+str(mc_mean))
print("mc_med: "+str(mc_med))
print("mc_std: "+str(mc_std))
```

The median for Metacritic_norm_round is lower than the mean because the score tends to distribute on the right part, but in the case of Fandango_Stars is the other way round.
The scores for Fandango_Stars distribute in a relatively high area, no score is less than 3.0, while the range of Metacritic_norm_round scores are much wider.
So the mean for Fandango_Stars is much higher, and standard deviation is much lower.

```python
x = np.random.uniform(-0.25, 0.25, movies.shape[0])
y = np.random.uniform(-0.25, 0.25, movies.shape[0])
plt.scatter(movies["Metacritic_norm_round"]+x, movies["Fandango_Stars"]+y)
plt.show()
```

```python
movies["fm_diff"] = np.abs(movies["Fandango_Stars"] - movies["Metacritic_norm_round"])
movies.sort_values("fm_diff", inplace=True, ascending=False)
print(movies.head(5)[["FILM", "Fandango_Stars", "Metacritic_norm_round"]])
```

```python
r_fm = stats.pearsonr(movies["Fandango_Stars"], movies["Metacritic_norm_round"])
print("r_value: " + str(r_fm))
```

The correlation is very low. Maybe one of them is not good.

```python
w, b, _, _, _ = stats.linregress(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
pred_3 = 3 * w + b
print(pred_3)
```

```python
pred_1 = 1 * w + b
print(pred_1)
pred_5 = 5 * w + b
print(pred_5)
```

```python
dx = np.random.uniform(-0.25, 0.25, movies.shape[0])
dy = np.random.uniform(-0.25, 0.25, movies.shape[0])
plt.scatter(movies["Metacritic_norm_round"]+dx, movies["Fandango_Stars"]+dy)
x1 = np.arange(1.0, 6.0)
y1 = x1 * w + b
plt.plot(x1, y1)
plt.xlim(1.0, 5.0)
plt.show()
```
